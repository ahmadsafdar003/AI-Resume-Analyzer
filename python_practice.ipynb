**Python Data-types**

**Numbers**

```python
print("Number printing like int, float")
x = 10        # int
y = 3.14      # float
z = 2 + 3j    # complex

print(x, type(x)) # <class 'int'>
print(y, type(y)) # <class 'float'>
print(z, type(z)) # <class 'complex'>

#We can also find the type like below
type(x)
#or
print(type(y), "\n")
```

**String**

```python
#string Datatype
print("String printing")
name = "Python"
print(name[0])        # P
print(name[-1])       # n
print(name[1:4])      # yth
print(name.upper())   # PYTHON
print(len(name))      # 6
print(name[1: 4])     # start from 1 and go through 4 means yth, y is located at 1 index and in range of 4 means less than 4 which is 3
```

**Multi Elements Data-types**

**LIST**

```python
#"1" List is An ordered, mutable collection that allows changing, inserting, deleting of elements during a program

print("\n")
fruits= ["apple", "banana", "orange", "banana"]
print(len(fruits))

fruits.append("grapes")  #to add something at the end
print(fruits)

fruits.remove("banana") #remove something but exact 1 argument with exact value not an index like here a banana
print(fruits)

del fruits[0]  #It's an opposite of remove here we can remove an element with the help of an index number.
print(fruits)

fruits.insert(1, "cherry") #at index 1
print(fruits)

fruits.pop(0) #Remove the last element

#print(fruits[0])
#print(fruits[1:3]) #The slice [start:end] means Start at index star (inclusive). Stop at index end (exclusive).
#print(fruits[0:]) #from index0 to onward
```

**TUPLE**

```python
#tuple is also multi-element data type that store a multiple vlaues, but non mutable means that changing inserting etc are not allowed like list

int = (2, 13, 13, 4, 5, 6434)
print(int)
int.count(10)  #we can do such operation in tuple but we can't change the elements etc
```

**SET**

```python
# Set is also multi-element data type, but here the indexes are not fixed, means
# the indexes of an element changes every time we run the program

set = {"apple", "banana", "banana", "cherry"}
print(set)
set.add("orange")
print(set)
set.remove("banana")
print(set)


#set.clear()
print(set, "\n")

print(len(set))

```
